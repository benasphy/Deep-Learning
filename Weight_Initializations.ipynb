- **Xavier Initialization**

```python
import numpy as np

n_in, n_out = 4, 3

# Xavier uniform
limit = np.sqrt(6 / (n_in + n_out))
W_xavier = np.random.uniform(-limit, limit, (n_out, n_in))

print("Limit:", limit)
print("Xavier Initialized Weights:\n", W_xavier)
print("Variance:", np.var(W_xavier))
```

- **HE Initialization**

```python
import numpy as np

n_in, n_out = 4, 3

# He normal initialization
std = np.sqrt(2 / n_in)
W_he = np.random.randn(n_out, n_in) * std

print("Standard deviation:", std)
print("He Initialized Weights:\n", W_he)
print("Variance:", np.var(W_he))
```
