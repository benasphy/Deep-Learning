```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_gradient_flow(initial_gradient=1.0, layers=50, mode='vanishing'):
    gradients = [initial_gradient]

    # Choose multiplication factor based on mode
    if mode == 'vanishing':
        factor = 0.8  # shrinks the gradient
    elif mode == 'exploding':
        factor = 1.2  # grows the gradient
    else:
        raise ValueError("Mode must be 'vanishing' or 'exploding'")

    # Simulate passing the gradient through layers
    for _ in range(1, layers):
        gradients.append(gradients[-1] * factor)

    # Plotting
    plt.figure(figsize=(10, 6))
    plt.plot(range(layers), gradients, marker='o')
    plt.yscale('log')  # Log scale to better see vanishing/exploding
    plt.xlabel('Layer')
    plt.ylabel('Gradient Magnitude (log scale)')
    plt.title(f'{mode.capitalize()} Gradient Simulation (factor={factor})')
    plt.grid(True, which="both", linestyle="--")
    plt.show()

    # Print final gradient
    print(f"Final gradient after {layers} layers: {gradients[-1]}")

# Example usage:

# Simulate vanishing gradient
simulate_gradient_flow(mode='vanishing')

# Simulate exploding gradient
simulate_gradient_flow(mode='exploding')
```
